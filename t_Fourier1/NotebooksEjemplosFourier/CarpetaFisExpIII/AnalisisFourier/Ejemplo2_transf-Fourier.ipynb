Transformada de Fourier de un paquete gaussiano:
$$f(t)=e^{-t^2}\cos(2\pi f*t)$$

```python
import numpy as np
import matplotlib.pyplot as plt

f=10 #Frecuencia de la señal

sample_freq=100 #Frecuencia de muestreo (Número de puntos o muestras de señal por unidad de tiempo)
dt=1/sample_freq #tiempo entre puntos de la señal.

t_start=0 #Tiempo de inicio de la señal
t_end=10 #Tiempo de finalización de la señal

shift=4 #Valor de tiempo que se va desplazar la señal a la derecha
#--------------------------------------------------------

#Se generan los puntos de la señal y se grafica.
time=np.arange(t_start,t_end,dt)
signal=np.exp(-(time-shift)**2)*np.cos(2*np.pi*f*(time-shift))
plt.plot(time,signal,'-')
plt.xlabel('Time')
plt.ylabel('amplitude')
plt.show()

#---------------------------------------------------------
#Cálculo de la transformada de Fourier
signal_FFT=np.fft.fft(signal/len(signal)) #Señal dividida entre el número de muestras
signal_FFT=signal_FFT[range(int(len(signal)/2))] #Toma solo la mitad del arreglo de valores

#Cálculo de las frecuencias correspondientes a cada valor de la transformada.
indexes=np.arange(len(signal_FFT)) #Crea el arreglo de índices 0, 1, 2,... para los valores del arreglo "signal_FT"
time_duration=len(signal)/sample_freq
frequencies=indexes/time_duration

plt.plot(frequencies,np.abs(signal_FFT))
plt.xlabel('frequency')
plt.ylabel('amplitude')
plt.show()
#-------------------------------------------------
```
