```python
import numpy as np
import matplotlib.pyplot as plt

sample_freq=1000 #Frecuencia de muestreo (Número de puntos o muestras de señal por unidad de tiempo)
dt=1/sample_freq #tiempo entre puntos de la señal.
t_start=0
t_end=1

#En este ejemplo simularemos tres pulsos gaussianos de diferentes alturas y el mismo ancho.
width=0.02 #Ancho de los pulsos gaussianos
pos1=0.25 #Posición (unidades de tiempo) de cada uno de los tres picos gaussianos
pos2=0.5
pos3=0.71

time=np.arange(t_start, t_end, dt)
signal=np.exp(-((time-pos1)**2/width**2))+2*np.exp(-((time-pos2)**2/width**2))+3*np.exp(-((time-pos3)**2/width**2))

plt.plot(time,signal)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
#Ahora calculamos la transformada de Fourier, pero incluyendo también las frecuencias negativas para poder comprobar la relación de Parseval.

signalFFT=signalFFT=np.fft.fftshift(np.fft.fft(np.fft.fftshift(signal)))
f_min=-sample_freq/2
f_max=sample_freq/2
df=sample_freq/len(signal)

frequencies=np.arange(f_min,f_max,df)

plt.plot(frequencies,np.abs(signalFFT))
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```

Ahora comprobaremos la relación de Parseval:
$\int |f(t)|^2dt=\int |f(\omega)|^2d\omega$, que llevada al caso discreto se convierte en:
$\sum_{n=o}^{N-1}|x_n|^2=\frac{1}{N}\sum_{k=o}^{N-1}|X_k|^2$, donde $X_k$ son los valores de la transformada de Fourier de los valores $x_n$.

```python
sum_signal=np.sum(np.abs(signal)**2)
sum_signalFFT=np.sum(np.abs(signalFFT)**2)

print("lado izquierdo:"+str(sum_signal))
print("lado derecho:"+str(sum_signalFFT/len(signal)))
```
