En este ejemplo calculamos la transformada de Fourier discreta para una señal formada mediante la superposición de dos señales tipo coseno de diferente frecuecia:
$$f_1(t)=\cos(2\pi f_1*t)$$
$$f_2(t)=\cos(2\pi f_2*t)$$
$$signal(t)=f_1(t)+f_2(t)$$

```python
import numpy as np
import matplotlib.pyplot as plt

f1=4 #Frecuencia de la primera señal
f2=25 #Frecuencia de la segunda señal

sample_freq=100 #Frecuencia de muestreo (Número de puntos o muestras de señal por unidad de tiempo)
dt=1/sample_freq #tiempo entre puntos de la señal.

t_start=0 #Tiempo de inicio de la señal
t_end=1 #Tiempo de finalización de la señal
```

```python
#Se generan los puntos de la señal y se grafica.
time=np.arange(t_start,t_end,dt)
signal=np.cos(2*np.pi*f1*time)+np.cos(2*np.pi*f2*time)
plt.plot(time,signal,'-')
plt.xlabel('Time')
plt.ylabel('amplitude')
plt.show()
```

```python
#Cálculo de la transformada de Fourier
signal_FFT=np.fft.fft(signal/len(signal)) #Ver https://numpy.org/doc/stable/reference/generated/numpy.fft.fft.html
#signal_FFT2=np.fft.fftshift(np.fft.fft(np.fft.fftshift(signal)/len(signal)))

plt.plot(np.abs(signal_FFT))
plt.show()
#plt.plot(signal_FFT2)
#plt.show()
#signal_FFT=np.fft.fft(signal) #Observar qué ocurre con la tranformada si no se hace la normalización anterior.
signal_FFT=signal_FFT[range(int(len(signal)/2))] #Toma solo la mitad del arreglo de valores para omitir las frecuencias negativas

##Estas líneas son solo para ilustrar el efecto de aplicar "numpy_array[range(n)]"
#test=np.arange(0,1,0.1)
#print(test)
#test=test[range(5)]
#print(test)

#Cálculo de las frecuencias correspondientes a cada valor de la transformada.
indexes=np.arange(len(signal_FFT)) #Crea el arreglo de índices 0, 1, 2,... para los valores del arreglo "signal_FT"
time_duration=len(signal)/sample_freq
frequencies=indexes/time_duration

plt.plot(frequencies,np.abs(signal_FFT))
plt.xlabel('frequency')
plt.ylabel('amplitude')
```

* Observar el efecto de acercar la frecuencia de muestreo a la frecuencia de las señales y asociarlo con el teorema de Nyquist.
